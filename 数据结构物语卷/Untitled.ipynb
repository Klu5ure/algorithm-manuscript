```python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList():
    def __init__(self):
        self.head = None
        self.size = 0
        
    def checkBorder(self, index):
        if index < 0 or index >= self.size:
            return False
        return True
    
    def append(self, data):
        node = Node(data)
        if not self.head:
            self.head = node
            self.size += 1
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = node
        self.size += 1
        
    def display(self):
        temp = self.head
        while temp:
            print(temp.data)
            temp = temp.next
            
    def get(self, index):
        if not self.checkBorder(index):
            return -1
        i = 0
        temp = self.head
        while i < index:
            if not temp:
                return
            temp = temp.next
            i += 1
        return temp.data
    
    def addAtHead(self, data):
        node = Node(data)
        node.next = self.head
        self.head = node
        self.size += 1
            
    def deleteAtIndex(self, index):
        if not self.checkBorder(index):
            return -1
        i = 0
        temp = self.head
        while i < index - 1:
            i += 1
            temp = temp.next
        temp.next = temp.next.next
        self.size -= 1
        
    def insert(self, index, data):
        if not self.checkBorder(index):
            return
        node = Node(data)
        i = 0
        temp = self.head
        if index == 0:
            self.addAtHead(data)
            return
        while i < index - 1:
            temp = temp.next
            i += 1
        node.next = temp.next
        temp.next = node
        self.size += 1
            
list = LinkedList()
list.addAtHead(2)
list.deleteAtIndex(1)
list.addAtHead(2)
list.addAtHead(7)
list.addAtHead(3)
list.addAtHead(2)
list.addAtHead(5)
list.addAtTail(5)
list.display()
list.get(5)
```

```python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class MyLinkedList:

    def __init__(self):
        self.head = Node(0)


    def get(self, index: int) -> int:
        node = self.head
        flag = 0
        while flag <= index and node:
            flag += 1
            node = node.next
        if not node:
            return -1
        return node.data


    def addAtHead(self, data: int) -> None:
        new_node = Node(data)
        new_node.next = self.head.next
        self.head.next = new_node


    def addAtTail(self, data: int) -> None:
        node = self.head
        while node.next:
            node = node.next
        node.next = Node(data)


    def addAtIndex(self, index: int, data: int) -> None:
        new_node = Node(data)
        node = self.head
        flag = 0
        while flag < index and node:
            node = node.next
            flag += 1
        if not node:
            return
        new_node.next = node.next
        node.next = new_node

    def display(self):
        temp = self.head
        while temp:
            print(temp.data)
            temp = temp.next

    def deleteAtIndex(self, index: int) -> None:
        node = self.head
        flag = 0
        while flag < index and node.next:
            node = node.next
            flag += 1
        if not node or not node.next:
            return
        node.next = node.next.next
        return
            
list = LinkedList()
list.addAtHead(2)
list.addAtHead(2)
list.addAtHead(7)
list.addAtHead(3)
list.addAtHead(2)
list.addAtHead(5)
list.addAtTail(5)
list.deleteAtIndex(0)
list.display()
# list.get(5)
```
