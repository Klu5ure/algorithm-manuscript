# 链表

```python
# 有头结点
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class MyLinkedList:

    def __init__(self):
        self.head = Node(0)


    def get(self, index: int) -> int:
        node = self.head
        flag = 0
        while flag <= index and node:
            flag += 1
            node = node.next
        if not node:
            return -1
        return node.data


    def addAtHead(self, data: int) -> None:
        new_node = Node(data)
        new_node.next = self.head.next
        self.head.next = new_node


    def addAtTail(self, data: int) -> None:
        node = self.head
        while node.next:
            node = node.next
        node.next = Node(data)


    def addAtIndex(self, index: int, data: int) -> None:
        new_node = Node(data)
        node = self.head
        flag = 0
        while flag < index and node:
            node = node.next
            flag += 1
        if not node:
            return
        new_node.next = node.next
        node.next = new_node

    def display(self):
        temp = self.head
        while temp:
            print(temp.data)
            temp = temp.next

    def deleteAtIndex(self, index: int) -> None:
        node = self.head
        flag = 0
        while flag < index and node.next:
            node = node.next
            flag += 1
        if not node or not node.next:
            return
        node.next = node.next.next
        return
            
list = LinkedList()
list.addAtHead(2)
list.addAtHead(2)
list.addAtHead(7)
list.addAtHead(3)
list.addAtHead(2)
list.addAtHead(5)
list.addAtTail(5)
list.deleteAtIndex(0)
list.display()
# list.get(5)
```

# 动态数组

```python
class ArrayList:

    def __init__(self):
        self.list = []

    def get(self, index: int) -> int:
        if index < 0 or index >= len(self.list):
            return -1
        return self.list[index]

    def addAtHead(self, val: int) -> None:
        self.list.insert(0, val)

    def addAtTail(self, val: int) -> None:
        self.list.append(val)

    def addAtIndex(self, index: int, val: int) -> None:
        if index < 0 or index > len(self.list):
            return -1
        self.list.insert(index, val)

    def deleteAtIndex(self, index: int) -> None:
        if index < 0 or index >= len(self.list):
            return -1
        self.list.pop(index)
```

# 链栈

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class MyStack:

    def __init__(self):
        self.top_node = None

    def push(self, x: int) -> None:
        node = Node(x)
        node.next = self.top_node
        self.top_node = node

    def pop(self) -> int:
        if self.empty():
            return -1
        node = self.top_node
        self.top_node = node.next
        return node.data

    def top(self) -> int:
        if self.empty():
            return -1
        return self.top_node.data

    def empty(self) -> bool:
        return self.top_node is None


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```

# 队列栈

```python
# 用列表模拟队列，只能用pop(0)

class MyStack:

    def __init__(self):
        self.pushQueue = []
        self.popQueue = []
        

    def push(self, x: int) -> None:
        self.pushQueue.append(x)

    def pop(self) -> int:
        if not self.popQueue:
            while len(self.pushQueue) > 1:
                self.popQueue.append(self.pushQueue.pop(0))
        res = self.pushQueue.pop()
        self.pushQueue, self.popQueue = self.popQueue, self.pushQueue
        return res 

    def top(self) -> int:
        if not self.popQueue:
            while len(self.pushQueue) > 1:
                self.popQueue.append(self.pushQueue.pop(0))
        res = self.pushQueue[0]
        self.popQueue.append(self.pushQueue.pop())
        self.pushQueue, self.popQueue = self.popQueue, self.pushQueue
        return res


    def empty(self) -> bool:
        return not self.popQueue and not self.pushQueue



# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```

# 链队列

```python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None

class MyQueue:

    def __init__(self):
        self.head = None 

    def push(self, x: int) -> None:
        node = self.head
        if not self.head:
            self.head = Node(x)
            return
        while node.next:
            node = node.next
        node.next = Node(x)

    def pop(self) -> int:
        node = self.head
        self.head = self.head.next
        return node.data


    def peek(self) -> int:
        return self.head.data


    def empty(self) -> bool:
        return self.head is None
```

# 栈队列

```python
class MyQueue:

    def __init__(self):
        self.pushStack = []
        self.popStack = []

    def push(self, x: int) -> None:
        self.pushStack.append(x)

    def pop(self) -> int:
        if not self.popStack:
            while self.pushStack:
                self.popStack.append(self.pushStack.pop())
        return self.popStack.pop()

    def peek(self) -> int:
        if not self.popStack:
            while self.pushStack:
                self.popStack.append(self.pushStack.pop())
        return self.popStack[-1]

    def empty(self) -> bool:
        return not self.pushStack and not self.popStack
```
